```python
#Importing pandas library for reading and using different functions with dataframe
import pandas as pd
```

```python
#Reading the dataset and displaying it
df = pd.read_csv('preprocessed_data.csv')
df
```

```python
#Summary Statistics
df.describe()
```

```python
#Correlation Matrix
numerical_columns = df.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()
correlation_matrix
```

```python
#Price and size have strong positive correlation 0.605

#Price and pricePerSquareMeter have strong positive correlation 0.702 showing that
#higher prices per square meter drive total price

#These features can be key predictors for price
#Price also has moderate correlations with numOfRooms 0.583 and population 0.409

#status, year and equipped are having weak correlations with price meaning they might not significantly affect price

#size, pricePerSquareMeter and numOfRooms should be primary predictors for regression model

#size and numOfRooms could have multicollinearity due to 0.654 correlation coefficient
#we can calculate VIF or Variance Inflation Factor to check for multicollinearity
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

#Preparation of attributes used
X = df[['size', 'pricePerSquareMeter', 'numOfRooms']]

#What is also useful is use of standardization, however this is not necessary
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Perform calculations for Variance Inflation Factor
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]
vif_data
```

```python
#Since multicollinearity is less than 5, no corrective actions are needed
```

```python
#Importing libraries for Ridge Regression
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#RIDGE REGRESSION
# We are selecting our features on which we performed VIF
#Since they do not have multicollinearity
selected_features = ['size', 'pricePerSquareMeter', 'numOfRooms']
X = df[selected_features]
y = df['price']

#Performing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Scaterplots for features with high correlation
plt.figure(figsize=(15, 5))
for i, feature in enumerate(selected_features):
    plt.subplot(1, 3, i + 1)
    sns.scatterplot(x=df[feature], y=df['price'])
    plt.title(f'Scatterplot: {feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
plt.tight_layout()
plt.show()

#Usage of ridge regression
#Splitting data between training and test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#Fitting the ridge regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

#Predicting
y_pred = ridge.predict(X_test)

#Evaluation by using MSE and R2
print("Ridge Regression Coefficients:", ridge.coef_)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))

#Visualization
coefficients = pd.DataFrame({
    'Feature': selected_features,
    'Coefficient': ridge.coef_
})
sns.barplot(x='Feature', y='Coefficient', data=coefficients)
plt.title('Ridge Regression Coefficients')
plt.show()
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

#Splitting data between training and testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

#Predicting with ridge
ridge_pred = ridge.predict(X_test)

#MSE, R2 evaulation for Ridge
print("Ridge Regression Results:")
print("Mean Squared Error (MSE):", mean_squared_error(y_test, ridge_pred))
print("R^2 Score:", r2_score(y_test, ridge_pred))
print("Coefficients:", ridge.coef_)

#Performing LASSO Regression
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

#Predicting with lasso
lasso_pred = lasso.predict(X_test)

#MSE, R2 evaulation for Lasso
print("\nLasso Regression Results:")
print("Mean Squared Error (MSE):", mean_squared_error(y_test, lasso_pred))
print("R^2 Score:", r2_score(y_test, lasso_pred))
print("Coefficients:", lasso.coef_)

#Visualization
coefficients = pd.DataFrame({
    'Feature': ['size', 'pricePerSquareMeter', 'numOfRooms'],
    'Ridge Coefficient': ridge.coef_,
    'Lasso Coefficient': lasso.coef_
})
coefficients = coefficients.melt(id_vars='Feature', var_name='Model', value_name='Coefficient')

sns.barplot(x='Feature', y='Coefficient', hue='Model', data=coefficients)
plt.title('Comparison of Ridge and Lasso Coefficients')
plt.show()
```

```python
#Lasso has slightly better MSE than Ridge
#Both models have same explanatory power with R2 score
#Minor differences, similr coefficients

#pricePerSquareMeter has largest impact on price, then sizen and numOfRooms with small but 
#meaningful impact on price

#keeping all features and creating model, ridge should be used
# to do some feature selection we could use lasso
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge, Lasso

# Define the parameter grid for alpha values
param_grid = {'alpha': [0.01, 0.1, 1, 10, 50, 100, 300, 400, 500, 1000, 3000, 4000, 5000, 8000]}

# Ridge Regression with Grid Search
ridge = Ridge()
ridge_grid = GridSearchCV(estimator=ridge, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
ridge_grid.fit(X_scaled, y)

# Best parameters and corresponding score for Ridge
print("Best alpha for Ridge Regression:", ridge_grid.best_params_['alpha'])
print("Best MSE for Ridge Regression:", -ridge_grid.best_score_)

# Lasso Regression with Grid Search
lasso = Lasso()
lasso_grid = GridSearchCV(estimator=lasso, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
lasso_grid.fit(X_scaled, y)

# Best parameters and corresponding score for Lasso
print("Best alpha for Lasso Regression:", lasso_grid.best_params_['alpha'])
print("Best MSE for Lasso Regression:", -lasso_grid.best_score_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Retrain Ridge Regression with best alpha
ridge_best = Ridge(alpha=300)
ridge_best.fit(X_scaled, y)
ridge_predictions = ridge_best.predict(X_scaled)

# Evaluate Ridge Regression
ridge_mse = mean_squared_error(y, ridge_predictions)
ridge_r2 = r2_score(y, ridge_predictions)
print("Ridge Regression - Best Alpha:")
print("MSE:", ridge_mse)
print("R^2 Score:", ridge_r2)

# Retrain Lasso Regression with best alpha
lasso_best = Lasso(alpha=5000)
lasso_best.fit(X_scaled, y)
lasso_predictions = lasso_best.predict(X_scaled)

# Evaluate Lasso Regression
lasso_mse = mean_squared_error(y, lasso_predictions)
lasso_r2 = r2_score(y, lasso_predictions)
print("\nLasso Regression - Best Alpha:")
print("MSE:", lasso_mse)
print("R^2 Score:", lasso_r2)
```

```python
import matplotlib.pyplot as plt

# Plot Ridge Predictions
plt.scatter(y, ridge_predictions, alpha=0.5, label='Ridge Predictions')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Ridge Regression - Actual vs Predicted')
plt.legend()
plt.show()

# Plot Lasso Predictions
plt.scatter(y, lasso_predictions, alpha=0.5, label='Lasso Predictions')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Lasso Regression - Actual vs Predicted')
plt.legend()
plt.show()
```
