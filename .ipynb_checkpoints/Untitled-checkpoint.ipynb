```python
import matplotlib.pyplot as plt

# Filter the data for the specified cities in 'cityShort'
cities_to_compare = ['Zenica', 'Sarajevo', 'Banja Luka', 'Grad Mostar', 'Bijeljina', 'Tuzla']
filtered_df = df[df['cityShort'].isin(cities_to_compare)]

# Group by 'cityShort' and count the number of entries (rows)
city_comparison = filtered_df['cityShort'].value_counts().reset_index()

# Rename columns for clarity
city_comparison.columns = ['cityShort', 'count']

# Create a bar plot to compare the counts
plt.figure(figsize=(10, 6))
plt.bar(city_comparison['cityShort'], city_comparison['count'], color=['blue', 'orange', 'green', 'red', 'grey', 'violet'])

# Add labels and title
plt.title('Comparison of Number of Entries between biggest cities')
plt.xlabel('City')
plt.ylabel('Number of Entries')

# Display the plot
plt.show()

```
