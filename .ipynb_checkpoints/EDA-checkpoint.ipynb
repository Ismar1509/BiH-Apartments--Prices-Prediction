```python
#Importing pandas library for reading and using different functions with dataframe
import pandas as pd
```

```python
#Reading the dataset and displaying it
df = pd.read_csv('preprocessed_data.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Correlation matrix for numerical columns
correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12, 8))

#Heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=True)
plt.title("Correlation Matrix Heatmap", fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#The strongest relationships in the dataset are between price and other size-related features
#size, pricePerSquareMeter, and numOfRooms.
```

```python
import matplotlib.pyplot as plt

#Comparison between Big Cities like Banja Luka, Zenica, Tuzla similar against Sarajevo
filtered_df = df[df['cityCat'].isin(['Big Cities', 'Sarajevo'])]

# Group by 'cityCat' and count the number of entries (rows)
city_comparison = filtered_df['cityCat'].value_counts().reset_index()
city_comparison.columns = ['cityCat', 'count']

# Create a bar plot to compare the counts
plt.figure(figsize=(8, 6))
plt.bar(city_comparison['cityCat'], city_comparison['count'], color=['blue', 'orange'])
plt.title('Comparison of Number of Apartments for selling: Big Cities vs Sarajevo')
plt.xlabel('City Category')
plt.ylabel('Number of Apartments')
plt.show()
```
