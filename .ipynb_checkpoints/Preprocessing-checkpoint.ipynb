```python
#Importing pandas library for reading and using different functions with dataframe
import pandas as pd
```

```python
#Reading the dataset and displaying it
df = pd.read_csv('BiH_Prices.csv')
df
```

```python
#Getting info about the data in dataset
df.info()
```

```python
df.describe()
```

```python
#Checking if we have Null values
df.isnull().sum()
```

```python
#We see that for 'size' and 'pricePerSquareMeter' we have 211 missing values, so we will try to input mean for missing values
#For population, we will update data with 2013 census on population in Bosnia and Herzegovina
```

```python
#Since the data is extracted from linkName and each property is in BAM, we can drop following columns, as they do not have any impact on predictions
new_df = df.drop(columns=['linkName', 'link', 'currencyKM'])
```

```python
new_df
```

```python
#Since 'status' shows if the apartment is either new or used, we can perform binarization for better understanding
new_df['status'] = new_df['status'].apply(lambda x: 1 if x == 'novo' else 0)
new_df
```

```python
new_df.info()
```

```python
#We want to impute mean price for missing values
#We imputed based on numOfRooms, so we calculated mean for apartments that have 3 rooms and added to missing values for apartments with 3 rooms
trosoban_data = new_df[new_df['numOfRooms'] == 'Trosoban (3)']
trosoban_mean_size = trosoban_data['size'].mean()
new_df.loc[(new_df['numOfRooms'] == 'Trosoban (3)') & (new_df['size'].isnull()), 'size'] = trosoban_mean_size
```

```python
new_df
```

```python
new_df[new_df['numOfRooms'] == 'Trosoban (3)']
```

```python
new_df
```

```python
new_df.info()
```

```python
#Similarly, mean price is calculated for 4 room apartments and added for 4 room apartments with missing values
cetverosoban_data = new_df[new_df['numOfRooms'] == 'Četverosoban (4)']
cetverosoban_mean_size = cetverosoban_data['size'].mean()
new_df.loc[(new_df['numOfRooms'] == 'Četverosoban (4)') & (new_df['size'].isnull()), 'size'] = cetverosoban_mean_size
new_df
```

```python
#Similar process for 2 room apartments
dvosoban_data = new_df[new_df['numOfRooms'] == 'Dvosoban (2)']
dvosoban_mean_size = dvosoban_data['size'].mean()
new_df.loc[(new_df['numOfRooms'] == 'Dvosoban (2)') & (new_df['size'].isnull()), 'size'] = dvosoban_mean_size
new_df
```

```python
new_df.info()
```

```python
#Process for 1 room apartment
jednosoban_data = new_df[new_df['numOfRooms'] == 'Jednosoban (1)']
jednosoban_mean_size = jednosoban_data['size'].mean()
new_df.loc[(new_df['numOfRooms'] == 'Jednosoban (1)') & (new_df['size'].isnull()), 'size'] = jednosoban_mean_size
new_df
```

```python
new_df.info()
```

```python
#For 1.5 room apartments
jednoiposoban_data = new_df[new_df['numOfRooms'] == 'Jednoiposoban (1.5)']
jednoiposoban_mean_size = jednoiposoban_data['size'].mean()
new_df.loc[(new_df['numOfRooms'] == 'Jednoiposoban (1.5)') & (new_df['size'].isnull()), 'size'] = jednoiposoban_mean_size
new_df
```

```python
#For 5 or more rooms apartments
petosoban_data = new_df[new_df['numOfRooms'] == 'Petosoban i više']
petosoban_mean_size = petosoban_data['size'].mean()
new_df.loc[(new_df['numOfRooms'] == 'Petosoban i više') & (new_df['size'].isnull()), 'size'] = petosoban_mean_size
new_df
```

```python
#No missing values for 'size'
new_df.info()
```

```python
#Checking missing values for pricePerSquareMeter column
na_values = new_df[new_df['pricePerSquareMeter'].isna()]
na_values
```

```python
#We performed calculation based on formula price/size for pricePerSquareMeter and applied to those entries that had missing values in pricePerSquareMeter
new_df['pricePerSquareMeter'] = new_df.apply(
    lambda row: row['price'] / row['size'] if pd.isna(row['pricePerSquareMeter']) else row['pricePerSquareMeter'], 
    axis=1
)
new_df
```

```python
#Checking if there are missing values in pricePerSquareMeter column
na_values = new_df[new_df['pricePerSquareMeter'].isna()]
na_values
```

```python
new_df.info()
```

```python
#Since 'equipped' column has 3 outcomes - equipped, half-equipped and non-equipped, we decided to map numerical values from 0 to 2
#2 for equipped, 1 for half-equipped and 0 for non-equipped
equipped_map = {
    'Namješten': 2,
    'Polunamješten': 1,
    'Nenamješten': 0
}
new_df['equipped'] = new_df['equipped'].map(equipped_map)
new_df
```

```python
new_df.info()
```

```python
#Rounding decimal values to 2 for size column
new_df['size'] = new_df['size'].round(2)
new_df
```

```python
new_df.info()
```

```python
#Converting dateNew column to datetime type
new_df['dateNew'] = pd.to_datetime(new_df['dateNew'], errors='coerce')
```

```python
#Extracting year and month from dateNew
new_df['year'] = new_df['dateNew'].dt.year
new_df['month'] = new_df['dateNew'].dt.month
new_df
```

```python
#Since we only need year and month, we can drop date and dateNew once we extracted year and month
new_df = new_df.drop(columns=['date', 'dateNew'])
new_df
```

```python
#Since we have numOfRooms and numOfRoom2 columns, we can drop numOfRooms, since numOfRooms2 is holding numerical values for number of rooms
#Thus it is more useful to use that column
new_df = new_df.drop(columns=['numOfRooms'])
new_df
```

```python
#Renaming column numOfRooms 2 to numOfRooms
new_df['numOfRooms'] = new_df['numOfRooms2']
new_df
```

```python
#Dropping numOfRooms2
new_df = new_df.drop(columns=['numOfRooms2'])
new_df
```

```python
new_df.info()
```

```python
#Dropping views and id columns since they are not necessary for our predictions
new_df = new_df.drop(columns=['views', 'id'])
new_df
```

```python
#Since all apartments are in Bosnia and Herzegovina, 'country' column is redundant
new_df = new_df.drop(columns=['country'])
new_df
```

```python
#We have names of the cities, thus latitude and longitude are redundant
new_df = new_df.drop(columns=['lat','long'])
new_df
```
